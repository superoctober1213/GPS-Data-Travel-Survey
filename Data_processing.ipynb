```python
import pandas as pd
df = pd.read_csv('jax gps.csv')
```

```python
# processing gps data from Jacksonville
new_df = pd.DataFrame(columns=['id, latitude, longitude, time'])

new_df = pd.DataFrame([{'id': row[1].split('|')[0],
                     'latitude': float(row[1].split('|')[1].strip('"')),
                     'longitude': float(row[1].split('|')[2].strip('"')),
                     'time': row[1].split('|')[9].strip('"')} for row in df.itertuples()])
```

```python
new_df
```

```python
selected_data = new_df.reset_index()
selected_data = selected_data.rename(columns={'index': 'ID'})
selected_data = selected_data.set_index('ID')
selected_data
```

```python
counts = selected_data.groupby('id').size()
counts = counts[counts > 100]
ids = counts.index.tolist()
filter_df = selected_data[selected_data['id'].isin(ids)]
filter_df
```

```python
filter_df['user_id'] = pd.factorize(filter_df['id'])[0]
```

```python
filter_df.to_csv('data.csv')
```
