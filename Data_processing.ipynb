```python
import pandas as pd
df = pd.read_csv('north fl.csv')
```

```python
df
```

```python
# processing gps data from Jacksonville
new_df = pd.DataFrame(columns=['user_id, latitude, longitude, flag, time'])

new_df = pd.DataFrame([{'user_id': row[1].split('|')[0],
                     'latitude': float(row[1].split('|')[1].strip('"')),
                     'longitude': float(row[1].split('|')[2].strip('"')),
                     'flag': int(row[1].split('|')[6].strip('"')),
                     'time': row[1].split('|')[9].strip('"')} for row in df.itertuples()])
```

```python
new_df
```

```python
new_df['user_id'] = pd.factorize(new_df['user_id'])[0]
new_df
```

```python
flags = new_df.flag

#full flags
I_dau = [((int(format(flag,'032b')[18:25:2])==0) & 
          (int(format(flag,'032b')[16])==1) &
          (int(format(flag,'032b')[14:16])==0) &
          (int(format(flag,'032b')[2:5])==0) &
          (int(format(flag,'032b')[8:12])==0)) for flag in flags]

#moderate_high flag
# I_dau = [((int(format(flag,'032b')[18:25:2])==0) & 
#           ((int(format(flag,'032b')[15])==1) |
#           (int(format(flag,'032b')[16])==1)) &
#           (int(format(flag,'032b')[2:5])==0) &
#           (int(format(flag,'032b')[8:12])==0)) for flag in flags]
```

```python
flag_df = new_df.copy()
flag_df['flag'] = I_dau
flag_df
```

```python
data = flag_df.drop(flag_df[flag_df['flag'] == False].index)
data
```

```python
selected_data = data.reset_index()
selected_data = selected_data.rename(columns={'index': 'ID'})
selected_data = selected_data.set_index('ID')
selected_data
```

```python
selected_data['time'] = pd.to_datetime(selected_data['time'])
sorted_data = selected_data.sort_values(by=['user_id', 'time'])
sorted_data
```
